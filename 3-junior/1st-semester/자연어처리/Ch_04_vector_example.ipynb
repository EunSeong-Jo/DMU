### 거리 구하기

```python
a = np.array([1,2])
b = np.array([2,2])
c = np.array([-3, -3])
```

```python
# 벡터 a의 길이 구하기
np.linalg.norm(a)
```

1. 유클리디안 거리

norm() 함수를 이용해서 두 벡터 사이의 유클리디안 거리 구하기

```python
print(np.linalg.norm(b-a))
print(np.linalg.norm(c-a))
```

scipy에서 distance 모듈에서 제공되는 함수를 이용한 유클리디안 거리 계산

```python
import scipy.spatial.distance as dst
```

```python
dst.euclidean(a,b)
```

```python
dst.euclidean(a,c)
```

2. 시티블록 거리

linalg 모듈의 norm() 함수를 이용한 L1-norm 거리 계산

```python
np.linalg.norm(c-a, 1)
```

scipy: 시티블록 거리 (L1 norm) 계산하기

```python
dst.cityblock(a,b)
```

```python
dst.cityblock(a,c)
```

3. Hamming distance

```python
d = np.array([1,0,0])
e = np.array([1,1,0])
```

```python
dst.hamming(d,e)
```

4. Jaccard distance

```python
dst.jaccard(d,e) # 확인
```

# Cosine similarity

#### 내적 (dot product)

```python
a = np.array([1,2])
b = np.array([2,2])
np.dot(a,b)
```

```python
# 코사인 유사도
np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
```

#### cosine distance

```python
dst.cosine(a,b)
```

```python
1-np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
```

# [doc_vectorization]

```python
# 텍스트 데이터 (말뭉치)
# 아래 TEXT 변수의 각 원소가 하나의 문서를 의미합니다.
# CountVectorizer나 TfidfVectorizer 클래스를 이용하기 위해서는
# 각 문서는 단어들로 구성된 하나의 문자열값으로 저장되어야 합니다.
TEXT = ['banana apple apple eggplant',
        'orange carrot banana eggplant',
        'apple carrot banana banana',
        'orange banana grape'
]
```

# CountVectorizer 이용하기

```python
from sklearn.feature_extraction.text import CountVectorizer # frequency based DTM

# min_df=2: 단어가 최소한 2개 이상의 문서에 등장
# max_df=0.8: 단어가 전체 문서의 80% 이상에서 등장하면 벡터에서 제외
# ngram_range=(1,1): 단어 단위의 토큰만 추출(유니그램)
tf_vectorizer = CountVectorizer(min_df=2, max_df=0.8, ngram_range=(1,1))
tf_features = tf_vectorizer.fit_transform(TEXT)
```

```python
print(tf_features)
```

```python
features = np.array(tf_features.todense())
```

```python
features
```

```python
# 첫 번째 문서의 벡터
features[0]
```

```python
features[1]
```

```python
# 단어들의 이름 확인하기
feature_names = tf_vectorizer.get_feature_names_out()
print(feature_names)
```

```python
import pandas as pd
df = pd.DataFrame(data=features, columns=feature_names)
df
```

문서들 간의 유사도 계산해 보기

```python
# 첫 번째 문서와 두 번째 문서 간 유클리디안 거리
print(np.linalg.norm(features[1]-features[0]))
```

```python
# 첫 번째 문서와 두 번째 문서 간 코사인 유사도
print(np.dot(features[0],features[1])/(np.linalg.norm(features[0])*np.linalg.norm(features[1])))
```

# TfidfVectorizer 이용하기

```python
from sklearn.feature_extraction.text import TfidfVectorizer # tf-idf based DTM
tfidf_vectorizer = TfidfVectorizer(min_df=2, max_df=0.8, ngram_range=(1,1))
tfidf_features = tfidf_vectorizer.fit_transform(TEXT)
```

```python
print(tfidf_features)
```

```python
tfidf_features = np.array(tfidf_features.todense())
tfidf_features
```

```python
# 첫 번째 문서의 벡터
tfidf_features[0]
```

```python
tfidf_feature_names = tfidf_vectorizer.get_feature_names_out()
df = pd.DataFrame(data=tfidf_features, columns=tfidf_feature_names)
df
```

```python
# 벡터의 길이 = 1
np.linalg.norm(tfidf_features[0])
```

```python
# 첫 번째 문서와 두 번째 문서 간의 유클리디안 거리 구하기
np.linalg.norm(tfidf_features[1]-tfidf_features[0])
```

```python
# 첫 번째 문서와 두 번째 문서 간 코사인 유사도
print(np.dot(tfidf_features[0],tfidf_features[1])/(np.linalg.norm(tfidf_features[0])*np.linalg.norm(tfidf_features[1])))
```
