```python
import torch
import numpy as np

def basic_tensor():
    print('텐서는 무엇인가')

    scalar = torch.tensor(5.0)
    print(f'스칼라 (0차원) : {scalar}, 모양 : {scalar.shape}')

    vector = torch.tensor([1.0, 2.0, 3.0])
    print(f'벡터 (1차원) : {vector}, 모양 : {vector.shape}')

    matrix = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
    print(f'행렬 (2차원) : {matrix}, 모양 : {matrix.shape}')

    print('텐서는 숫자들을 n차원 배열로 정리한 것')

def tensor_operations():
    print('텐서로 계산하기')

    a = torch.tensor([1.0, 2.0, 3.0])
    b = torch.tensor([4.0, 5.0, 6.0])

    print(f'a = {a}')
    print(f'b = {b}')
    print(f'a + b = {a+b}')
    print(f'a * b = {a*b}')
    # 내적 : 벡터들의 방향이 얼마나 유사한가
    # 값이 클 수록 유사함
    print(f'a @ b = {a @ b}')
```
