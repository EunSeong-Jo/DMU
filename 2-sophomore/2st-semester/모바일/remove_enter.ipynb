```python
import os

def remove_all_whitespace_from_file(input_filename, output_filename):
    """
    텍스트 파일에서 줄바꿈, 띄어쓰기, 탭 문자를 포함한 모든 공백 문자를 제거하고 새 파일에 저장합니다.
    """
    try:
        with open(input_filename, 'r', encoding='utf-8') as infile:
            content = infile.read()

        # 줄바꿈(\n), 띄어쓰기( ), 탭(\t) 문자를 모두 제거
        processed_content = content.replace('\n', '') \
                                   .replace(' ', '') \
                                   .replace('\t', '')
        
        # Windows 환경의 경우 캐리지 리턴(\r)도 제거하는 것이 좋습니다.
        # processed_content = content.replace('\n', '').replace(' ', '').replace('\t', '').replace('\r', '')


        with open(output_filename, 'w', encoding='utf-8') as outfile:
            outfile.write(processed_content)

        print(f"✅ 모든 공백 문자 제거 완료!")
        print(f"   -> 원본 파일: '{input_filename}'")
        print(f"   -> 결과 파일: '{output_filename}'")

    except FileNotFoundError:
        print(f"❌ 오류: 파일 '{input_filename}'을(를) 찾을 수 없습니다.")
    except Exception as e:
        print(f"❌ 처리 중 오류 발생: {e}")

# --- 사용 예시 ---
input_file = './announcement.txt'
output_file = 'output.txt'

remove_all_whitespace_from_file(input_file, output_file)
```
