### 24년 1회

```python
# 다음 Python 코드에 알맞는 출력값을 작성하시오.

a = ["Seoul", "Kyeonggi", "Incheon", "Daejun", "Daegu", "Pusan"]
str01 = "S"

# 리스트 내부의 6개 값을 차례로 i에 할당
for i in a:
  # 각 i의 [1]인덱스를 더해서 기존 str01에 덮어씀
  # 첫번째 i : Seoul (S + e)
  # 두번쨰 i : Kyeonggi (Se + y)
  # 세번쨰 i : Incheon (Sey + n)
  # 네번쨰 i : Daejun (Seyn + a)
  # 다섯번쨰 i : Daegu (Seyna + a)
  # 마지막 i : Pusan (Seynaa + u)
  str01 = str01 + i[1]
# Seynaau
print(str01)
```

### 24년 2회

```python
# 다음 Python 코드에 알맞는 출력값을 작성하시오.

def func(x, y):
	result = 0
	for i in range(len(x)):
		sub = x[i:i+len(y)]
		if sub == y:
			result += 1
	return result

str = "abdcabcabca"
p1 = "ca"
p2 = "ab"

# ab + str 문자열 "ca" 개수
# ca + str 문자열 "ab" 개수
print(f'ab{func(str, p1)}', f'ca{func(str, p2)}')
```

예상문제

```python
def count_pattern(text, pattern):
    count = 0
    for i in range(len(text)):
        substring = text[i:i+len(pattern)]
        if substring == pattern:
            count += 1
    return count

string = "abababcabab"
pattern1 = "ab"
pattern2 = "ba"

print(f'ab{count_pattern(string, pattern2)}', f'ba{count_pattern(string, pattern1)}')
```

### 24년 3회

```python
def func(lst):
    # 리스트 길이의 절반만큼 반복해서 리스트의 순서를 바꿈
    for i in range(len(lst) // 2):
        # i번째 값과 뒤에서 i번째 값을 서로 바꿈.
        # i=0일 때: lst[0]과 lst[-1]을 교환
        # i=1일 때: lst[1]과 lst[-2]를 교환
        lst[i], lst[-i-1] = lst[-i-1], lst[i]

lst = [1, 2, 3, 4, 5, 6]

# [6, 5, 4, 3, 2, 1]
func(lst)

# lst[::2]는 짝수 인덱스(0, 2, 4)
# lst[1::2]는 홀수 인덱스(1, 3, 5)
# sum() 함수로 합을 구한 뒤 차이를 출력
# (6 + 4 + 2) - (5 + 3 + 1) = 12 - 9 = 3
print(sum(lst[::2]) - sum(lst[1::2]))
```

예상문제

```python
def reverse_list(arr):
    for i in range(len(arr) // 2):
        arr[i], arr[-i-1] = arr[-i-1], arr[i]

numbers = [2, 4, 6, 8, 10, 12]
reverse_list(numbers)

# numbers는 [12, 10, 8, 6, 4, 2]가 됨
# 홀수 인덱스의 합에서 짝수 인덱스의 합을 뺌
print(sum(numbers[1::2]) - sum(numbers[::2]))
```
