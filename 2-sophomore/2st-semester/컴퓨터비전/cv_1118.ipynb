```python
from pixellib.semantic import semantic_segmentation
import cv2 as cv

seg = semantic_segmentation()
seg.load_ade20k_model("deeplabv3_xception65_ade20k.h5")

img_fname = 'image.png'

seg.segmentAsAde20k(img_fname, output_image_name="image9-5.png")

info1, img_segmented1 = seg.segmentAsAde20k(img_fname)
info2, img_segmented2 = seg.segmentAsAde20k(img_fname, overlay=True)

cv.imshow('origin image', cv.imread(img_fname))
cv.imshow('seg1', img_segmented1)
cv.imshow('seg2', img_segmented2)

cv.waitKey()
cv.destroyAllWindows()
```

```python
from pixellib.instance import instance_segmentation
import cv2 as cv

seg = instance_segmentation()
seg.load_model("mask_rcnn_coco.h5")

img_fname = 'image.png'

info1, img_segmented = seg.segmentImage(img_fname, show_bboxes=True)

cv.imshow('origin image', img_segmented)

cv.waitKey()
cv.destroyAllWindows()
```
