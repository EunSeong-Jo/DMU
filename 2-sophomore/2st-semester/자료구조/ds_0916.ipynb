```python
# 숫자 입력을 받아서 짝수이면 "짝수입니다."를 출력하고 홀수이면 "홀수입니다."를 출력하는 프로그램
num = int(input("숫자를 입력하세요: "))
if num % 2 == 0:
    print("짝수입니다.")
else:
    print("홀수입니다.")
```

```python
# 성적을 입력받아서 90점 이상이면 A, 80점 이상이면 B, 70점 이상이면 C, 60점 이상이면 D, 그 외에는 F를 출력하는 프로그램

score = int(input("성적을 입력하세요: "))
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

```python
def find_min_max(A):
    min =A[0]
    max =A[0]
    for i in range(1, len(A)):
        if max < A[i]:
            min = i
        if min > A[i]:
            max = i
    return min(A), max(A)
```
