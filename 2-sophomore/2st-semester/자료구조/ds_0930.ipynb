```python
class ArrayList:
    def __init__(self, capacity=10):
        self.capacity = capacity
        self.array = [None] * self.capacity
        self.size = 0

    def isEmpty(self):
        return self.size == 0
    
    def isFull(self):
        return self.size == self.capacity
    
    def getEnty(self, pos):
        if pos < 0 or pos >= self.size:
            raise IndexError("Index out of bounds")
        return self.array[pos]

    def insert(self, pos, e):
        if not self.isFull() and 0 <= pos <= self.size:
            for i in range(self.size, pos, -1):
                self.array[i] = self.array[i - 1]
            self.array[pos] = e
            self.size += 1
        else:
            pass

    def delete(self, pos):
        if not self.isEmpty() and 0 <= pos < self.size:
            e = self.array[pos]
            for i in range(pos, self.size - 1):
                self.array[i] = self.array[i + 1]
            self.size -= 1
            return e
        else:
            pass
```

```python
class ArrayStack :
    def __init__( self, capacity):         # 생성자 정의
        self.capacity = capacity           # 용량(고정)
        self.array = [None]*self.capacity   # 요소들을 저장할 배열
        self.top = -1                     # 스택 상단의 인덱스

    # 코드 1.2b: 스택 클래스의 연산들
    def isEmpty( self ) :
       return self.top == -1

    def isFull( self ) :
       return self.top == self.capacity-1

    def push( self, item ):
        if not self.isFull() :
            self.top += 1
            self.array[self.top] = item
        else: pass              # overflow 예외는 처리하지 않았음

    def pop( self ):
        if not self.isEmpty():
            self.top -= 1
            return self.array[self.top+1]
        else: pass              # underflow 예외는 처리하지 않았음

    def peek( self ):
        if not self.isEmpty():
            return self.array[self.top]
        else: pass              # underflow 예외는 처리하지 않았음

    def __str__(self ) :
        return str(self.array[0:self.top+1][::-1])

    def size( self ) : return self.top+1
```

```python
my_stack = ArrayStack(5)

# 데이터 삽입
my_stack.push('a')
my_stack.push('b')
my_stack.push('c')

# 최상단 항목 확인
print(my_stack.peek())

# 데이터 추출
print(my_stack.pop())

# 스택이 비었는지 확인
print(my_stack.isEmpty())
```

```python
s = ArrayStack(10)
msg = input("문자열 입력 : ")

print(msg)

for c in msg:
    s.push(c)
print(s)

while not s.isEmpty():
    print(s.pop(), end='')
```

```python
Class Car:
    def __init__(self, model, color, speed=0):
        self.model = model
        self.color = color
        self.speed = speed

    def speed_up(self, increment):
        self.speed += increment
        print(f"The car has accelerated. Current speed: {self.speed} km/h")

    def speed_down(self, decrement):
        self.speed = max(0, self.speed - decrement)
        print(f"The car has decelerated. Current speed: {self.speed} km/h")
```
