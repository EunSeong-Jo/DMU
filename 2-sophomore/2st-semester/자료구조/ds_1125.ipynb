```python
# 어떤 이진트리의 높이가 4라면 최소와 최대로 몇 개의 노드를 가질 수 있을는가?
'''
최소 : 4개
최대 : 15개 (2^n -1)
'''

# 이진트리에서 간선의 수 E와 노드의 수 N의 관계는?
'''
N = E + 1
'''

# 이진트리를 배열로 표현했을 때 인덱스가 15번인 노드의 왼쪽 자식과 오른쪽 자식의 위치는?
# (단, 0번 인덱스는 사용하지 않음.)
'''
왼쪽 자식 : 30
오른쪽 자식 : 31
'''

# 이진트리의 링크 표현법에서 노드수가 N인 이진트리의 None 링크의 수는?
'''
2n - (N - 1)
None 링크의 수 : N + 1
'''
```

```python
class TNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = None
        self.right = None
    def isLeaf(self):
        return self.left is None and self.right is None


    def preorder(n):
        if n is not None:
            print(n.data, end=' ')
            n.preorder(n.left)
            n.preorder(n.right)

    def inorder(n):
        if n is not None:
            n.inorder(n.left)
            print(n.data, end=' ')
            n.inorder(n.right)

    def postorder(n):
        if n is not None:
            n.postorder(n.left)
            n.postorder(n.right)
            print(n.data, end=' ')
```

```python
root = TNode(1, None, None)
root.left = TNode(2, None, None)
root.right = TNode(3, None, None)
root.left.left = TNode(4, None, None)
root.left.right = TNode(5, None, None)
root.right.left = TNode(6, None, None)
root.right.right = TNode(7, None, None)

print("Preorder: ", end='')
TNode.preorder(root)
print("\nInorder: ", end='')
TNode.inorder(root)
print("\nPostorder: ", end='')
TNode.postorder(root)
```

```python
def count_leaf(n) :
    if n is None : return 0
    elif n.isLeaf() : return 1
    else : return count_leaf(n.left) + count_leaf(n.right)

def count_node(n) :
    if n is None : 
        return 0
    else : 
        return 1 + count_node(n.left) + count_node(n.right)

def calc_height(n) :
    if n is None : return 0
    hLeft = calc_height(n.left)
    hRight = calc_height(n.right)
    if (hLeft > hRight) : return hLeft + 1
    else: return hRight + 1
```
