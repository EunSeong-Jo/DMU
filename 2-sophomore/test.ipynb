```python
# 문제 3: 지뢰찾기

m, n = map(int, input().split())
board = []

for x in range(m):
    board.append(list(input()))

# (-1,-1) (-1, 0) (-1, 1)
# (0,-1)  (0, 0)  (0, 1)
# (1,-1)  (1, 0)  (1,1)
dx = [-1, -1, -1, 0, 0, 1, 1, 1]
dy = [-1, 0, 1, -1, 1, -1, 0, 1]

result = []
for i in range(m):
    row = ""
    for j in range(n):
        if board[i][j] == '@':
            row += '@'
        else:
            count = 0
            for k in range(8):
                # 현재 위치의 주변 8개의 값
                ni, nj = i + dx[k], j + dy[k]
                # 매트릭스 내부에 존재하고 주변에 지뢰가 있을떄
                if 0 <= ni < m and 0 <= nj < n and board[ni][nj] == '@':
                    count += 1
            # 문자열로 변환하여 합쳐짐 방지
            row += str(count)
    print(row)
```

```python
# 문제 5: 휴대폰 떨어뜨리기 (Egg Drop Problem)

n, k = map(int, input().split())

num_of_trial = [[0] * (k + 1) for z in range(n + 1)]

for i in range(1, n + 1):
    num_of_trial[i][1] = 1
    num_of_trial[i][0] = 0

for j in range(1, k + 1):
    num_of_trial[1][j] = j

for i in range(2, n + 1):
    for j in range(2, k + 1):
        num_of_trial[i][j] = float('inf')
        for x in range(1, j + 1):
            res = 1 + max(num_of_trial[i - 1][x - 1], num_of_trial[i][j - x])
            num_of_trial[i][j] = min(num_of_trial[i][j], res)

print(num_of_trial[n][k])
```

```python
# 문제 5: 휴대폰 떨어뜨리기 (Egg Drop Problem)

n, k = map(int, input().split())

# 휴대폰 n개, k층 건물의 최소 시행 횟수
num_of_trial = [[0] * (k + 1) for z in range(n + 1)]

# 1층인 경우
for i in range(1, n + 1):
    num_of_trial[i][1] = 1

# 휴대폰이 1개인 경우
for j in range(1, k + 1):
    num_of_trial[1][j] = j

# 휴대폰과 층이 2개 이상인 경우
for i in range(2, n + 1):
    for j in range(2, k + 1):
        num_of_trial[i][j] = k + 1
        
        # x층에서 떨어뜨리는 모든 경우를 시도
        for x in range(1, j + 1):
            # 휴대폰이 깨지는 경우 (휴대폰 - 1, 층 - 1)
            # 휴대폰이 살아남는 경우 (휴대폰은 그대로, 남은 위층만 탐색)
            current_trial = 1 + max(num_of_trial[i - 1][x - 1], num_of_trial[i][j - x])
            num_of_trial[i][j] = min(num_of_trial[i][j], current_trial)

print(num_of_trial[n][k])
```

```python
# 문제 4: 그래프 컬러링 (4색 문제)
# 함수 없이 스택을 사용한 백트래킹

# 입력 받기
color_num = int(input())

# 인접 행렬 입력
adj = []
while True:
    line = input()
    if line == '':
        break
    adj.append(line)

# 노드 수 (0번 행/열은 보초이므로 제외)
node_count = len(adj) - 1

# 각 노드에 칠한 색을 저장 (0은 아직 안 칠함)
coloring = []
for i in range(node_count + 1):
    coloring.append(0)

# 결과 카운트
count = 0

# 현재 노드와 시도할 색 번호
current_node = 1
current_color = 1

while True:
    # 현재 노드에서 시도할 색이 남아있는 경우
    if current_color <= color_num:
        
        # 현재 색이 유효한지 확인
        flag = True
        for other in range(1, node_count + 1):
            is_adjacent = (adj[current_node][other] == '1')
            same_color = (coloring[other] == current_color)
            if is_adjacent and same_color:
                flag = False
                break
        
        if flag:
            # 색 칠하기
            coloring[current_node] = current_color
            
            # 모든 노드를 다 칠했는지 확인
            if current_node == node_count:
                # 하나의 완성된 경우 찾음
                count = count + 1
                # 색 지우고 다음 색 시도
                coloring[current_node] = 0
                current_color = current_color + 1
            else:
                # 다음 노드로 이동
                current_node = current_node + 1
                current_color = 1
        else:
            # 유효하지 않으면 다음 색 시도
            current_color = current_color + 1
    
    else:
        # 현재 노드에서 시도할 색이 없음 -> 이전 노드로 돌아가기 (백트래킹)
        coloring[current_node] = 0
        current_node = current_node - 1
        
        # 1번 노드보다 앞으로 가면 종료
        if current_node < 1:
            break
        
        # 이전 노드에서 다음 색부터 시도
        current_color = coloring[current_node] + 1
        coloring[current_node] = 0

print(count)
```

```python
# 문제 4: 그래프 컬러링

color_num = int(input())

adj = []
while True:
    line = input()
    if line == '':
        break
    adj.append(line)

node_count = len(adj) - 1

coloring = []
for i in range(node_count + 1):
    coloring.append(0)

count = 0
current_node = 1
current_color = 1

while current_node >= 1:
    
    if current_color <= color_num:
        # 현재 색이 유효한지 확인
        flag = True
        for other in range(1, node_count + 1):
            if adj[current_node][other] == '1' and coloring[other] == current_color:
                flag = False
                break
        
        if flag:
            coloring[current_node] = current_color
            
            if current_node == node_count:
                # 완성된 경우 찾음
                count = count + 1
                coloring[current_node] = 0
                current_color = current_color + 1
            else:
                # 다음 노드로 이동
                current_node = current_node + 1
                current_color = 1
        else:
            current_color = current_color + 1
    
    else:
        # 백트래킹: 이전 노드로 돌아가기
        current_node = current_node - 1
        if current_node >= 1:
            current_color = coloring[current_node] + 1
            coloring[current_node] = 0

print(count)
```

```python
# 색 가지 수
color_num = int(input())

# 인접 행렬 입력
graph=','.join(iter(input,''))
adj=[list(map(int,i)) for i in graph.split(',')]

# 보초(0)를 제외한 전체 노드 수
node_count = len(adj) - 1

# 칠한 색상 저장
coloring = []
for i in range(node_count + 1):
    coloring.append(0)

count = 0
# 시작 노드 번호
current_node = 1
# 시작 색 번호
current_color = 1

# 1번 노드 이전으로 돌아가면 모든 탐색 완료
while current_node >= 1:
    # 아직 시도할 색이 남아있는 경우
    if current_color <= color_num:
        # 인접한 노드 중 같은 색이 있는지 확인
        flag = True
        for other in range(1, node_count + 1):
            if adj[current_node][other] == 1 and coloring[other] == current_color:
                flag = False
                break
        
        if flag:
            # 현재 노드에 색 칠하기
            coloring[current_node] = current_color
            
            # 마지막 노드까지 칠한 경우, 경우의 수 +1
            if current_node == node_count:
                count = count + 1
                # 다른 경우를 찾기 위해 현재 색상을 지우고 다음 색상 시도
                coloring[current_node] = 0
                current_color = current_color + 1
            # 마지막 노드까지 도착하지 못한 경우
            else:
                # 다음 노드로 이동, 색상은 1번부터 시도
                current_node = current_node + 1
                current_color = 1
        else:
            # 현재 색상이 유효하지 않으면 다음 색상 시도
            current_color = current_color + 1
    
    else:
        # 모든 색상을 시도했지만 유효한 색상이 없음
        # 이전 노드로 돌아가서 다른 색 시도
        current_node = current_node - 1
        if current_node >= 1:
            # 이전 노드에서 칠했던 색의 다음 색부터 시도
            current_color = coloring[current_node] + 1
            coloring[current_node] = 0

print(count)
```
