#### 📘 JSP & Java 예상 출제 포인트 정리 (코드 빈칸형 대비용)

##### 🟦 JSP 관련 출제 포인트

| 출제 포인트                  | 코드 예시 |
|-----------------------------|-----------|
| JSP page 지시자 선언         | `<%@ page contentType="text/html; charset=UTF-8" pageEncoding="UTF-8" %>` |
| JSP import 지시자           | `<%@ page import="java.util.List" %>` |
| `request.getParameter()`     | `<% String userId = request.getParameter("user_id"); %>` |
| `request.getRemoteAddr()`    | `<% String ip = request.getRemoteAddr(); %>` |
| `request.setCharacterEncoding()` | `<% request.setCharacterEncoding("utf-8"); %>` |
| `session.setAttribute()`     | `<% session.setAttribute("userId", userId); %>` |
| `session.getAttribute()`     | `<% String user = (String)session.getAttribute("userId"); %>` |
| `session.invalidate()`       | `<% session.invalidate(); %>` |
| `response.sendRedirect()`    | `<% response.sendRedirect("/"); %>` |
| `<jsp:include>` 사용         | `<jsp:include page="include_menu.jsp" />` |
| JSP 표현식 `<%= %>`         | `<p>아이디: <%= userId %></p>` |
| HTML form method/form 태그  | `<form method="post" action="login-submit.jsp">` |
| `required` 속성             | `<input type="text" name="user_id" required />` |
| `<input type="hidden">` 사용 | `<input type="hidden" name="id" value="<%= board.getId() %>"/>` |
| `type="submit"`, `type="reset"` 차이 | `<button type="submit">`, `<button type="reset">` |
| JS 리다이렉트 방식           | `<script>location.href = 'bbs-list.jsp';</script>` |
| 삭제 확인 confirm           | `onclick="return confirm('삭제하시겠습니까?')"` |
| label과 input 연결           | `<label for="userId">아이디</label>` + `<input id="userId" ... />` |

##### 🟧 Java / DAO / Service 관련 출제 포인트

| 출제 포인트 | 코드 예시 |
|-------------|-----------|
| DTO 빌더 패턴 사용 | `MemberDto.builder().userId("test").userName("홍길동").build();` |
| JDBC 드라이버 로딩 | `Class.forName("com.mysql.cj.jdbc.Driver");` |
| DB 연결 | `Connection conn = DriverManager.getConnection(url, id, pwd);` |
| DbConnector 유틸 사용 | `Connection conn = DbConnector.getConnection();` |
| SQL 준비 | `PreparedStatement ps = conn.prepareStatement("SELECT * FROM member");` |
| 값 바인딩 | `ps.setString(1, member.getUserId());` |
| 쿼리 실행 | `ResultSet rs = ps.executeQuery();` |
| 쿼리 실행 (업데이트용) | `int affected = ps.executeUpdate();` |
| 결과 반복 처리 | `while (rs.next()) { ... }` |
| 리스트 생성 및 반환 | `List<MemberDto> list = new ArrayList<>();` |
| Stream 필터링 조회 | `list.stream().filter(e -> e.getId() == id).findFirst().orElse(null);` |
| try-catch 예외 처리 | `try { ... } catch (Exception e) { e.printStackTrace(); }` |
| Service에서 Dao 호출 | `BoardDao boardDao = new BoardDao();` |
| 메서드 boolean 결과 | `return result;` |
| 제네릭 문법 | `List<BoardDto>` |
| 클래스 implements 선언 | `public class MemberServiceImpl implements MemberService` |

##### 🟥 SQL / 데이터베이스 관련 출제 포인트

| 출제 포인트 | 예시 |
|-------------|------|
| 기본 SELECT 쿼리 | `SELECT * FROM member;` |
| 조건 SELECT | `SELECT * FROM board WHERE id = ?;` |
| INSERT | `INSERT INTO member (user_id, user_name, password) VALUES (?, ?, ?);` |
| UPDATE | `UPDATE member SET user_name = ?, password = ? WHERE user_id = ?;` |
| DELETE | `DELETE FROM member WHERE user_id = ?;` |
| AUTO_INCREMENT | `id BIGINT AUTO_INCREMENT PRIMARY KEY` |
| DEFAULT now() 사용 | `create_dt DATETIME DEFAULT now()` |
| 정렬 조건 | `ORDER BY update_dt DESC, create_dt DESC` |
| 컬럼 타입 구분 | `VARCHAR`, `TEXT`, `DATETIME`, `INT` 등 |

##### 🟨 기타 놓치기 쉬운 출제 포인트

| 출제 포인트 | 예시 |

| HTML 테이블 기본 구조 | `<table> <thead> <tbody> ... </tbody> </table>` |

##### ✅ 총정리: 출제 유형 정리표

| 유형 | 주요 예시 | 빈칸 위치 |
|------|-----------|------------|
| 내장 객체 사용 | `request`, `session`, `response` | 객체명, 메서드명 |
| SQL 쿼리 구문 | `SELECT`, `INSERT`, `UPDATE`, `DELETE` | 명령어, 컬럼명, 조건절 |
| JDBC 흐름 | `prepareStatement`, `executeQuery` 등 | 메서드 이름, SQL 문자열 |
| HTML 폼 처리 | `<input>`, `<form>`, `<button>` | 속성 값, 타입 |
| 자바 흐름 | Service → Dao → DB → DTO | 호출 순서, 메서드 명 |
| 유틸/로직 | `DateUtils`, `DbConnector`, 빌더 패턴 | 메서드, 클래스명 |

---

🎯 **활용 팁**  
- 위 표에서 나열된 코드 중 `메서드 이름`, `객체명`, `속성` 등이 빈칸으로 나올 수 있음  
- `DAO → Service → JSP` 흐름을 항상 염두에 둘 것  
- `form 전송 → 파라미터 수신 → DTO로 구성 → DAO 저장`의 흐름도 암기해두기

### 🕒 로그인 날짜 출력 기능 - JSP 시험 대비 정리

##### ✅ 1. 핵심 개념

| 요소 | 설명 | 출제 포인트 |
|------|------|--------------|
| `Date` | 현재 날짜 및 시간 객체 생성 | `new Date()` |
| `SimpleDateFormat` | 날짜 출력 형식 지정 클래스 | `new SimpleDateFormat("패턴")` |
| `session.setAttribute()` | 세션에 로그인 시간 저장 | 키/값 이름 |
| `session.getAttribute()` | 세션에서 시간 불러오기 | 키 일치 여부 |
| JSP 표현식 `<%= ... %>` | 세션값을 화면에 출력 | 출력 구조 |

---

##### 📌 로그인 처리 후 세션에 날짜 저장 (login-submit.jsp)

```jsp
<%@ page import="java.util.Date" %>
<%@ page import="java.text.SimpleDateFormat" %>

<%
    Date now = new Date();
    SimpleDateFormat sdf = new SimpleDateFormat("yyyy-MM-dd HH:mm:ss");

    session.setAttribute("loginTime", sdf.format(now));
%>

<%
    String loginTime = (String) session.getAttribute("loginTime");
%>
<% if (loginTime != null) { %>
    <p style="text-align:right;">로그인 시간: <%= loginTime %></p>
<% } %>
