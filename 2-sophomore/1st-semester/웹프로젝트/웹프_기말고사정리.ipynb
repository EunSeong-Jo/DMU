#### 📘 JSP & Java 예상 출제 포인트 정리

##### 🟦 JSP 관련 출제 포인트

| 출제 포인트 | 코드 예시 | 기능 설명 |
|-------------|-----------|------------|
| page 지시자 선언 | `<%@ page contentType="text/html; charset=UTF-8" pageEncoding="UTF-8" %>` | JSP 페이지 인코딩을 설정한다. |
| import 지시자 | `<%@ page import="java.util.List" %>` | JSP에서 사용할 Java 클래스를 불러온다. |
| `getParameter()` | `<% String userId = request.getParameter("user_id"); %>` | 클라이언트가 보낸 파라미터 값을 읽어온다. |
| `getRemoteAddr()` | `<% String ip = request.getRemoteAddr(); %>` | 접속자의 IP 주소를 가져온다. |
| `setCharacterEncoding()` | `<% request.setCharacterEncoding("utf-8"); %>` | 요청의 문자 인코딩을 UTF-8로 설정한다. |
| `setAttribute()` | `<% session.setAttribute("userId", userId); %>` | 세션에 데이터를 저장한다. |
| `getAttribute()` | `<% String user = (String)session.getAttribute("userId"); %>` | 세션에서 저장된 값을 읽어온다. |
| `invalidate()` | `<% session.invalidate(); %>` | 세션을 초기화하고 로그아웃 처리한다. |
| `sendRedirect()` | `<% response.sendRedirect("/"); %>` | 클라이언트를 다른 URL로 이동시킨다. |
| `<jsp:include>` | `<jsp:include page="include_menu.jsp" />` | JSP 페이지의 일부를 포함시킨다. |
| 표현식 출력 | `<p>아이디: <%= userId %></p>` | 변수 값을 HTML로 출력한다. |
| form 태그 | `<form method="post" action="login-submit.jsp">` | 사용자의 입력 데이터를 서버로 전송한다. |
| `required` 속성 | `<input type="text" name="user_id" required />` | 해당 입력란이 반드시 채워져야 함을 지정한다. |
| hidden input | `<input type="hidden" name="id" value="<%= board.getId() %>"/>` | 사용자에게 보이지 않게 데이터를 전송한다. |
| 버튼 타입 | `<button type="submit">`, `<button type="reset">` | 전송 또는 초기화 동작을 실행한다. |
| JS 이동 처리 | `<script>location.href = 'bbs-list.jsp';</script>` | 자바스크립트를 이용해 페이지를 이동한다. |
| 삭제 확인 | `onclick="return confirm('삭제하시겠습니까?')"` | 삭제 전 사용자 확인 팝업을 띄운다. |
| label-for | `<label for="userId">아이디</label>` | 라벨과 입력칸을 연결해 클릭 시 입력칸이 활성화된다. |

##### 🟧 Java / DAO / Service 출제 포인트

| 출제 포인트 | 코드 예시 | 기능 설명 |
|-------------|-----------|------------|
| 빌더 패턴 | `MemberDto.builder().userId("test").build();` | DTO 객체를 간결하게 생성한다. |
| 드라이버 로딩 | `Class.forName("com.mysql.cj.jdbc.Driver");` | MySQL JDBC 드라이버를 로딩한다. |
| DB 연결 | `DriverManager.getConnection(...)` | DB와 연결된 커넥션을 만든다. |
| 공용 DB 연결 | `DbConnector.getConnection();` | 유틸 클래스로부터 커넥션을 가져온다. |
| SQL 준비 | `prepareStatement("SELECT ...")` | 실행할 SQL 문을 미리 컴파일한다. |
| 값 바인딩 | `ps.setString(1, member.getUserId());` | SQL 문에 값을 바인딩한다. |
| SELECT 실행 | `ps.executeQuery();` | SELECT 문을 실행하고 결과를 받는다. |
| INSERT/UPDATE 실행 | `ps.executeUpdate();` | DB에 데이터 변경을 요청한다. |
| 결과 처리 | `while (rs.next()) { ... }` | 결과 집합을 반복하여 읽는다. |
| 리스트에 추가 | `list.add(member);` | 객체들을 리스트에 담는다. |
| Stream 필터 | `.stream().filter(...).findFirst()` | 조건에 맞는 데이터를 필터링한다. |
| try-catch | `try { ... } catch { ... }` | 예외 상황을 처리한다. |
| DAO 호출 | `new BoardDao().addBoard(...);` | DB 접근 객체를 이용해 동작을 수행한다. |
| boolean 결과 | `return result;` | 성공/실패 결과를 반환한다. |
| 제네릭 사용 | `List<MemberDto>` | 리스트 안의 객체 타입을 명시한다. |
| 인터페이스 구현 | `implements MemberService` | 인터페이스를 실제 구현한다. |

##### 🟥 SQL / DB 쿼리 출제 포인트

| 출제 포인트 | 예시 | 기능 설명 |
|-------------|------|------------|
| 전체 조회 | `SELECT * FROM member;` | 모든 행을 조회한다. |
| 조건 조회 | `SELECT * FROM board WHERE id = ?;` | 조건에 맞는 행만 조회한다. |
| 데이터 삽입 | `INSERT INTO member ...` | 테이블에 새 데이터를 추가한다. |
| 데이터 수정 | `UPDATE member SET ...` | 특정 행의 데이터를 수정한다. |
| 데이터 삭제 | `DELETE FROM member ...` | 특정 데이터를 삭제한다. |
| 자동 증가 | `AUTO_INCREMENT` | 기본키를 자동으로 증가시킨다. |
| 등록일 기본값 | `create_dt DATETIME DEFAULT now()` | 현재 시간을 기본값으로 설정한다. |
| 정렬 쿼리 | `ORDER BY ... DESC` | 결과를 내림차순 정렬한다. |
| 컬럼 타입 | `VARCHAR`, `TEXT`, `DATETIME`, `INT` | 각각 문자열, 긴 글, 시간, 숫자 타입이다. |

##### ✅ 요약: 출제 유형 정리표

| 유형 | 예시 | 빈칸 위치 |
|------|------|------------|
| JSP 내장객체 | `request`, `session`, `response` | 메서드 또는 객체 이름 |
| SQL | `SELECT`, `INSERT`, `UPDATE` 등 | 명령어, 컬럼명 |
| JDBC 흐름 | `ps.setString()`, `executeQuery()` | 메서드 이름 |
| 폼 입력 | `<input>`, `<form>` | 속성값 |
| Java 호출 흐름 | DAO → Service → JSP | 메서드명 |
| 유틸 클래스 | `DbConnector`, `DateUtils` | 메서드명, 클래스명 |

---

##### 🎯 시험 팁

- 빈칸은 보통 `객체 이름`, `메서드`, `속성`, `SQL 명령어`에서 나온다.
- 흐름을 묻는 문제는 `"폼 제출 → request 처리 → DTO → DAO"` 구조를 익히자.
- 내장 객체 이름은 정확히 암기하자 (`request`, `session`, `response` 등).


### 🕒 로그인 날짜 출력 기능 - JSP 시험 대비 정리

##### ✅ 1. 핵심 개념

| 요소 | 설명 | 출제 포인트 |
|------|------|--------------|
| `Date` | 현재 시간을 가져오는 기본 클래스 | `new Date()` 사용 위치 |
| `SimpleDateFormat` | 날짜를 지정한 문자열 형식으로 변환 | `"yyyy-MM-dd HH:mm:ss"`와 같은 형식 지정 |
| `session.setAttribute()` | 포맷된 날짜를 세션에 저장 | 키 이름 `"loginTime"` 정확히 기억 |
| `session.getAttribute()` | 세션에 저장된 시간을 꺼내오기 | 키 `"loginTime"`과 일치해야 정상 작동 |
| JSP 표현식 `<%= ... %>` | 세션의 값을 HTML에 출력 | `<%= loginTime %>` 구조 외우기 |

---

##### 📌 로그인 처리 후 세션에 날짜 저장 (예: login-submit.jsp 내에서 활용)

```jsp
<%@ page import="java.util.Date" %>
<%@ page import="java.text.SimpleDateFormat" %>

<%
    Date now = new Date();
    SimpleDateFormat sdf = new SimpleDateFormat("yyyy-MM-dd HH:mm:ss");

    session.setAttribute("loginTime", sdf.format(now));
%>

<%
    String loginTime = (String) session.getAttribute("loginTime");
%>
<% if (loginTime != null) { %>
    <p style="text-align:right;">로그인 시간: <%= loginTime %></p>
<% } %>
