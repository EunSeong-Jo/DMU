```python
df1 = pd.DataFrame({
    'sub' : ['AI', 'Python', 'Java'],
    'score' : [80, 77, 99]
})

df2 = pd.DataFrame({
    'sub' : ['AI', 'Python', 'C'],
    'book' : [30000, 28000, 25000]
})
```

```python
df1.merge(df2)
```

```python
df1.merge(df2).shape
```

```python
df1.merge(df2).shape[1]
```

---

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='left')['book'][1]
```

---

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='outer').iloc[2, 0]
```
