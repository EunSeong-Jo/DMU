## 시리즈 고급 인덱싱

```python
age = pd.Series([9, 17, 32, 45, 65, 98], index=list('abcdef'), name='age')
age
```

```python
age[ [1, 3, 5] ]
```

```python
age[ ['a', 'c'] ]
```

```python
age['f'] # 항목을 반환
```

```python
age[ ['f'] ] # 항목을 갖는 시리즈를 반환
```
