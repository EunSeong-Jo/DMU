```python
df_raw = pd.DataFrame({'var1' : [1, 2, 1],
           'var2' : [2, 3, 2],
           'var3' : [10, 20, 30]})

df_raw
```

```python
df_new = df_raw.copy()

df_new
```

```python
id(df_new) == id(df_raw)
```

```python
df_new.rename(columns={'var2' : 'v2'})
```

```python
df_new
```

```python
df_new.rename(columns={'var2' : 'v2'}, inplace=True)
```

```python
df_new
```

```python
df_new.rename(index={0:'x', 1:'y', 2:'z'}, inplace=True)

df_new
```
