```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# 데이터 불러오기
lifesat = pd.read_csv("./taxi_data.csv")

# 입력 변수와 타겟 변수 설정
x = lifesat[["distance(km)"]].values
y = lifesat[["price(1000)"]].values

# 산점도 그리기
lifesat.plot(kind='scatter', grid=True, x="distance(km)", y="price(1000)")
plt.xlabel("Distance (km)")
plt.ylabel("Price (1000)")

# 선형 회귀 모델 학습
model = LinearRegression()
model.fit(x, y)

# 회귀선 그리기
x_range = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)  # x 범위 설정
y_pred = model.predict(x_range)  # 예측 값 계산
plt.plot(x_range, y_pred, color='red', linewidth=2)  # 회귀선 추가

# 그래프 범위 설정
plt.axis([0, 100, 0, 100])
plt.show()

# 특정 거리(50km) 예측
x_new = [[50]]
pred = model.predict(x_new)

print(pred)
```
