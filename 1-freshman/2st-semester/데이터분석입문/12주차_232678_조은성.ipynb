```python
f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

male = []
female = []


name = '제주특별자치도'

for row in data :
    if name in row[0] :
        for i in range(3, 104) :
            male.append(int(row[i].replace(',', '')))
        for i in range(106, 207) :
            female.append(int(row[i].replace(',', '')))
        break

f.close()

plt.plot(male, label='male', color='b')
plt.plot(female, label='female', color='r')

plt.xlabel('Population')
plt.ylabel('Age')
plt.grid(True)
plt.legend()
plt.show()
```
