```python
df = pd.read_csv("age.csv", encoding="cp949", index_col=0)

df.head(3)
```

```python
df.dtypes
```

```python
df = df.replace(",", "", regex=True)

df.head(3)
```

```python
df = df.apply(pd.to_numeric)

df.dtypes
```

```python
df = df[df[df.columns[0]] != 0]
```

```python
df = df.div(df[df.columns[0]], axis="index")

df.head(3)
```

```python
del df["2024년09월_계_총인구수"], df["2024년09월_계_연령구간인구수"]                             

df.head(3)
```

```python
name = input("지역 이름 입력 : ")

a = df.index.str.contains(name)

df2 = df[a]

df2
```

```python
import matplotlib.pyplot as plt

name = input("지역 이름 입력 : ")

a = df.index.str.contains(name)

df2 = df[a]

# plt.rcParams["font.family"] = 

df2.T.plot()
plt.show()
```

```python
x = df.sub(df2.iloc[0], axis=1)

x.head(3)
```

```python
import numpy as np

y = np.power(x, 2)
z = y.sum(axis=1)

z
```

```python
i = z.sort_values().index[:6]

i
```

```python
# plt.rc("font", family = )

plt.rcParams["axes.unicode_minus"] = False

df.loc[i].T.plot(figsize = (10, 6))

plt.xlabel("age")
plt.ylabel("people")

plt.grid(True)
plt.show()
```
