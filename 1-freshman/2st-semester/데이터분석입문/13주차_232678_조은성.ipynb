```python
import csv
import matplotlib.pyplot as plt

f = open('./subwaytime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
next(data)

result = []

max_people = 0
max_station = ''
t = int(input('몇 시의 승차인원이 가장 많은 역? : '))

for row in data :
    for i in range(4, 52):
        row[i] = row[i].replace(',', '')
        
    row[4:] = map(int, row[4:])
    # 승차 부분이 인덱스 4번 항목부터 존재함
    # 승차 부분이 인덱스 짝수번호에 존재 함으로 *2
    # 탑승시간 인덱스 번호와 승차 시간 인덱스 번호를 맞추기 위해 -4
    people = row[2 * t - 4]
    if people > max_people:
        max_people = people
        max_station = row[3] + '(' + row[1] + ')'

f.close()

print(max_station, ':', max_people, '명')
```

```python
# 시간대별 최대 승차 역 이름 및 승차 인원 막대그래프로 시각화



f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
next(data)

max_people = [0] * 24
max_station = [''] * 24

for row in data :
    for i in range(4, 52):
        row[i] = row[i].replace(',', '')
        
    row[4:] = map(int, row[4:])
    for j in range(24):
        people = row[2 * j + 4]
        if people > max_people[j]:
            max_people[j] = people
            max_station[j] = row[3] + f'({str((j + 4) % 24)})'

f.close()

plt.rc('font', family='Malgun Gothic')

plt.bar(range(24), max_people)
plt.xticks(range(24), max_station, rotation=90)
plt.grid(True)
plt.show()
```

```python
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)
next(data)

s_in = [0] * 24
s_out = [0] * 24

for row in data :
    for i in range(4, 52):
        row[i] = row[i].replace(',', '')

    row[4:] = map(int, row[4:])
    for i in range(24) :
        s_in[i] += row[4 + i * 2]
        s_out[i] += row[5 + i * 2]

f.close()

plt.rc('font', family='Malgun Gothic')
plt.title('시간대별 승하차 인원')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
plt.grid(True)
plt.xticks(range(24), range(4, 28))
plt.show()
```

```python
f = open('subwayfee.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

max_station = ['' for _ in range(4)]
max_people = [0 for _ in range(4)]

label = ['유임승차','유임하차','무임승차','무임하차']

for row in data:
    for i in range(4, 8):
        row[i] = int(row[i].replace(',', ''))
        if row[i] > max_people[i-4]:
            max_people[i-4] = row[i]
            max_station[i-4] = row[3] + ' ' + row[1]

f.close()

for i in range(4):
    print(f'{label[i]}: {max_station[i]}, {max_people[i]} 명')
```

```python
import numpy as np

f = open('age.csv', encoding='cp949')
data = csv.reader(f)

name = input('지역 이름을 입력 : ')
for row in data:
    row[2:] = map(lambda s : s.replace(',', ''), row[2:])
    if name in row[0]:
        home = np.array(row[3:], dtype = int)
        break

f.close()

plt.rc('font', family='Malgun Gothic')
plt.plot(home)
plt.title(f'{name} 인구수')
plt.grid(True)
plt.xlabel('age')
plt.ylabel('person')
plt.show()
```

```python
f = open('age.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

name = input('지역 이름을 입력 : ')
for row in data:
    row[2:] = map(int, map(lambda s : s.replace(',', ''), row[2:]))
    if name in row[0]:
        home = np.array(row[3:], dtype = int) / row[2]
    if '화촌면' in row[0]:
        homeB = np.array(row[3:], dtype = int) / row[2]

f.close()

plt.rc('font', family='Malgun Gothic')
plt.plot(home)
plt.plot(homeB)
plt.title(f'{name} 인구수')
plt.grid(True)
plt.xlabel('age')
plt.ylabel('person')
plt.legend()
plt.show()
```

```python
f = open('age.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
data = list(data)

min_val = 1
result_name = ''
result = 0

name = input('지역 이름을 입력 : ')

for row in data:
    row[2:] = map(int, map(lambda s : s.replace(',', ''), row[2:]))
    if name in row[0]:
        home = np.array(row[3:], dtype = int) / row[2]
    
for row in data:
    if name not in row[0]:
        if row[2] == 0:
            continue
        
        away = np.array(row[3:], dtype=int)/row[2]
        s = np.sum((home - away)**2)

        if s < min_val:
            min_val = s
            result_name = row[0]
            result = away

f.close()

plt.rc('font', family='Malgun Gothic')
plt.plot(home)
plt.plot(result)
plt.title(f'{name} == {result_name}')
plt.grid(True)
plt.xlabel('age')
plt.ylabel('person')
plt.legend()
plt.show()
```
