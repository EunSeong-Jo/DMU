```python
f = open('./gender.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

male = []
female = []

for row in data:
    if '신도림' in row[0]:
        for i in row[3:104]:
            male.append(int(i))
        for i in row[106:]:
            female.append(int(i))

f.close()

plt.barh(range(len(male)), male, color='b', label='male')
plt.barh(range(len(female)), female, color='r', label='female')

plt.xlabel('population')
plt.ylabel('age')

plt.grid(True)
plt.legend()
plt.show()
```

```python
f = open('./gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

male = []
female = []

name = input('지역이름 입력 : ') 

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            i = i.replace(',', '')
            male.append(-int(i))
        for i in row[106:]:
            i = i.replace(',', '')
            female.append(int(i))

f.close()
plt.rc('font', family = 'Malgun Gothic')
plt.rc('axes', unicode_minus = False)

plt.barh(range(len(male)), male, color='b', label='male')
plt.barh(range(len(female)), female, color='r', label='female')

plt.title(f'{name} 성별 인구 분포')
plt.xlabel('population')
plt.ylabel('age')

plt.grid(True)
plt.legend()
plt.show()
```

```python
f = open('./gender.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

# name = input('지역이름 입력 : ')
name = '제주'

for row in data:
    if name in row[0]:
        male = 0
        female = 0
        for i in row[3:104]:
            i = i.replace(',', '')
            male = male + int(i)
        for i in row[106:]:
            i = i.replace(',', '')
            female = female + int(i)
        break

f.close()

labels = ['male','female']
colors = ['skyblue', 'pink']

plt.pie([male, female], labels=labels, autopct ='%.1f%%', colors=colors)
plt.title(name + '성별 비율')
plt.axis('equal')
plt.show()
```

```python
import numpy as np

plt.style.use('default')

x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c=colors)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8), ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
fig, ax = plt.subplots()

for color in ['tab:blue', 'tab:orange', 'tab:green']:
    n = 75
    x, y = np.random.rand(2, n)
    scale = 300.0 * np.random.rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color, alpha=0.3, edgecolors='red')

ax.legend()
ax.grid(True)
plt.show()
```

```python
n = 100
xmin, xmax, ymin, ymax, zmin, zmax = 0, 20, 0, 30, 0, 50
cmin, cmax = 0, 2

xs = np.array([(xmax - xmin) * np.random.random_sample() + xmin for i in range(n)])
ys = np.array([(ymax - ymin) * np.random.random_sample() + ymin for i in range(n)])
zs = np.array([(zmax - zmin) * np.random.random_sample() + zmin for i in range(n)])
color = np.array([(cmax - cmin) * np.random.random_sample() + cmin for i in range(n)])

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c=color, marker='o', s=15, cmap='Greens')

plt.show()
```
