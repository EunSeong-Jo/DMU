```python
import pandas as pd

# index_col : 선택된 열을 인덱스로 불러옴
# thousands : 1000자리 숫자를 ,(콤마) 구분없이 표시
pop = pd.read_csv("202311popmonth.csv", encoding='cp949', index_col=0, thousands=',')
pop
```

```python
pop.columns = ['총인구수', '세대수', '세대당인구', '남자인구수', '여자인구수', '남여비율']
pop.head()
```

```python
# pop.index : ['전국  (1000000000)', '서울특별시  (1100000000)', '부산광역시  (2600000000)', ... ]
# pop.index[0] : '전국  (1000000000)'
# pop.index[0].split() : ['전국', '(1000000000)']
# pop.index[0].split()[0] : '전국'

pop.index = [pop.index[i].split()[0] for i in range(len(pop))]
pop
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# scatterplot : 산점도
sns.scatterplot(pop['총인구수'])

# ticks: 축 설정
# rotation : 회전
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pop1 = pop.sort_values('총인구수')

sns.barplot(x = pop1.index, y = '총인구수', data = pop1)
plt.xticks(rotation=90)
plt.show()
```
